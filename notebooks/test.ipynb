```python


import sympy as sp

x = sp.Rational(3, 4)
print("3/4 is rational? \t\t", x.is_rational)
print("3/4 is integer? \t\t", x.is_integer)

r = sp.sqrt(2)
print("sqrt(2) is rational? \t", r.is_rational)
print("sqrt(2) is real? \t\t", r.is_real)
print("sqrt(2) approx: \t\t", sp.N(r, 12))
```

```python
from fractions import Fraction


def long_division_digits(p, q, n_digits=12):
    digits = []
    p = p % q
    for _ in range(n_digits):
        p *= 10
        digits.append(p // q)
        p = p % q
    return digits

print("1/7 ->",long_division_digits(1, 7))
print("1/4 ->",long_division_digits(1, 4))

x1 = Fraction(36, 99)
x2 = Fraction(4, 44)
print("x1=", x1)
print("x2=", x2)
print("x1 + x2=", x1 + x2)

print(1 % 4)
print(1 // 4)
```
