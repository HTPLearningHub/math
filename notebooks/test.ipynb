```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 400)
y = np.abs(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 3.5))

ax1.plot(x, y)
ax1.set_title("y = |x|  (distance from 0)")
ax1.axhline(0, color="gray", lw=0.8)
ax1.axvline(0, color="gray", lw=0.8)
ax1.grid(True, alpha=0.3)

ax2.hlines(0, -4, 4, color="black")
ax2.hlines(0, -1, 2, color="tab:blue", lw=6)
ax2.plot([-1, 2], [0, 0], "o", color="tab:blue")
ax2.set_title("the interval [-1, 2]")
ax2.set_ylim(-1, 1)
ax2.set_yticks([])
plt.tight_layout()
plt.show()
```
